```python
from bing_image_downloader import downloader

downloader.download("dog", limit=100)
```

```python
import cv2
import os

# Define the directory path where the images are located
directory = "/content/dataset/dog"

# Loop through all the files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Read the image
        img_path = os.path.join(directory, filename)
        img = cv2.imread(img_path)

        # Convert the image to grayscale
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Save the grayscale image with a new filename
        new_filename = "gray_" + filename
        new_img_path = os.path.join(directory, new_filename)
        cv2.imwrite(new_img_path, gray_img)

        print(f"Converted {filename} to grayscale: {new_filename}")
```
